# NEC(6701)株価情報
NEC(6701)の [**HP**](https://finance.yahoo.co.jp/quote/6701.T/history) から株価情報の最初のページのみ Pandas で取得
![株価情報](./Pandas-Exercises1.png)

```python
# ライブラリのインポート
import pandas as pd
from datetime import datetime as dt

# リクエスト送信とデータ取得
url = 'https://finance.yahoo.co.jp/quote/6701.T/history'
data = pd.read_html(url, header=0)

# 欠損値(NaN)を削除
data[0].dropna(inplace=True)

# 日付型のフィールド「修正日付」を追加
data[0]['修正日付'] = [dt.strptime(val, "%Y年%m月%d日") for val in data[0]['日付']]

# 不要なフィールド「日付」を削除
data[0].drop('日付', axis=1, inplace=True)

# フィールド「修正日付」をインデックス指定
data[0].set_index('修正日付', inplace=True)

# フィールド「修正日付」をキーに昇順にソート
data[0].sort_index(ascending=True, inplace=True)
```
